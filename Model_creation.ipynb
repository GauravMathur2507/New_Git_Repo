## Creating and saving the model

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
diabetes = load_diabetes()
diabetes
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(diabetes.data,diabetes.target,test_size=0.3,random_state=0)
```

```python
LR_model = LinearRegression()
```

```python
LR_model.fit(X_train,y_train)
```

```python
LR_model.predict(X_test)
```

```python
score = LR_model.score(X_test,y_test)
round(score,2)
```

## Saving the model

```python
import pickle
```

```python
#creating a pickle file
with open("pickle_diabetes_model.pkl","wb") as file:
    pickle.dump(LR_model,file)
```

```python
#open a pickle file
with open("pickle_diabetes_model.pkl","rb") as f:
    pickled_model_file = pickle.load(f)
```

```python
pickled_model_file.score(X_test,y_test)
```
